# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#Get the total rides for each city type
ride_count_type = pyber_data_df.groupby('type').count()['ride_id']
ride_count_type
```

```python
# 2. Get the total drivers for each city type
driver_total_type = city_data_df.groupby('type').sum()['driver_count']
driver_total_type
```

```python
#  3. Get the total amount of fares for each city type
# 2. Get the total drivers for each city type
fares_total_type = pyber_data_df.groupby('type').sum()['fare']
fares_total_type
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_type = fares_total_type / ride_count_type
average_fare_type
```

```python
# 5. Get the average fare per driver for each city type. 
driver_average_fare = fares_total_type / driver_total_type
driver_average_fare
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
    'Total Rides': ride_count_type,
    'Total Drivers': driver_total_type,
    'Total Fares': fares_total_type,
    'Average Fare per Ride': average_fare_type,
    'Average Fare per Driver': driver_average_fare})
pyber_summary_df
    
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df['Total Rides']= pyber_summary_df['Total Rides'].map("{:,}".format)
pyber_summary_df['Total Drivers']= pyber_summary_df['Total Drivers'].map("{:,}".format)
pyber_summary_df['Total Fares']= pyber_summary_df['Total Fares'].map("${:,.2f}".format)
pyber_summary_df['Average Fare per Ride']= pyber_summary_df['Average Fare per Ride'].map("${:,.2f}".format)
pyber_summary_df['Average Fare per Driver']= pyber_summary_df['Average Fare per Driver'].map("${:,.2f}".format)
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
#Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
pyber_city_date_df = pyber_data_df.groupby(['date', 'type']).sum()['fare']
pyber_city_date_df
```

```python
# Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
pyber_city_date_df = pyber_city_date_df.reset_index()
pyber_city_date_df
```

```python
# Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_city_date_df = pyber_city_date_df.pivot(index='date', columns = 'type', values = 'fare')
```

```python
pyber_city_date_df.head(10)
```

```python
# Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
date_range_df = pyber_city_date_df.loc['2019-01-01':'2019-04-29']
date_range_df
```

```python
# Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
date_range_df.index = pd.to_datetime(date_range_df.index)
```

```python
# Check that the datatype for the index is datetime using df.info()
date_range_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pyber_2019_weeks_df = date_range_df.resample('W').sum()
pyber_2019_weeks_df
```

```python
#Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style

# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
fig, ax = plt.subplots()
labels = ['Rural', 'Urban', 'Suburban']
months= ['Jan', 'Feb', 'Mar', 'Apr', 'May']
x_values = np.arange(len(list(pyber_2019_weeks_df.index.values)))
x_ticks =np.arange(len(list(pyber_2019_weeks_df.index.values)), step = 4.0)
y_values = pyber_2019_weeks_df
ax.plot(x_values, y_values, label = labels)
ax.set_xlabel('2019')
ax.set_ylabel('Fare($)')
ax.set_title('Total Fare By City Type')
ax.set_xticks(x_ticks, months)
ax.legend()
fig.set_size_inches(15,6)

plt.savefig("analysis/PyBer_fare_summary.png")
```

```python
#creating data frame for number of rides per week in city type
rides_city_date_df = pyber_data_df.groupby(['date', 'type']).count()['ride_id']
rides_city_date_df
```

```python
# Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
rides_city_date_df = rides_city_date_df.reset_index()
rides_city_date_df
```

```python
# Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
rides_city_date_df = rides_city_date_df.pivot(index='date', columns = 'type', values = 'ride_id')
```

```python
# Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
date_range_rides_df = rides_city_date_df.loc['2019-01-01':'2019-04-29']
date_range_rides_df
```

```python
# Create a new DataFrame using the "resample()" function by week 'W' .
pyber_2019_rides_df = date_range_rides_df.resample('W').sum()
pyber_2019_rides_df
```

```python
#plotting the resample rides DataFrame on a multiline graph

# Import the style from Matplotlib.
from matplotlib import style

# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
fig, ax = plt.subplots()
labels = ['Rural', 'Urban', 'Suburban']
months= ['Jan', 'Feb', 'Mar', 'Apr', 'May']
x_values = np.arange(len(list(pyber_2019_rides_df.index.values)))
x_ticks =np.arange(len(list(pyber_2019_rides_df.index.values)), step = 4.0)
y_values = pyber_2019_rides_df
ax.plot(x_values, y_values, label = labels)
ax.set_xlabel('2019')
ax.set_ylabel('Rides per Week)')
ax.set_title('Total Rides By City Type')
ax.set_xticks(x_ticks, months)
ax.legend()
fig.set_size_inches(15,6)

plt.savefig("analysis/PyBer_rides_summary.png")
```
